```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('max_columns', 50)
```

```python
q3 = pd.read_csv('data\Q3_5_Trip_Purpose.txt')
q5 = pd.read_csv('data\Q5 Dest Purpose.txt')
q7 = pd.read_csv('data\Q7 Egress Mode.txt')
q8 = pd.read_csv('data\Q8_Fare_Instrument.txt')
q9 = pd.read_csv('data\Q9_Transit_Benefit.txt')
q10 = pd.read_csv('data\Q10 Residence.txt')
q11 = pd.read_csv('data\Q11_Vehicles.txt')
q14 = pd.read_csv('data\Q14_Race.txt')
q16 = pd.read_csv('data\Q16_English.txt')
q17 = pd.read_csv('data\Q17_Income.txt')
q18 = pd.read_csv('data\Q18_Age.txt')
```

```python
wt = pd.read_csv('Projects\metro\Weekday_2014_2015-05-18.txt')
wt.head()
```

```python
#SELECT * FROM WT WHERE ROUTE = "16Y"

y = wt[wt['Route']=="16Y"]
```

```python
y.columns.values
```

```python
y.tail()
```

```python
a1 = pd.merge(y, q3, how='left', left_on='q3', right_on='Q3_5')
a = pd.merge(a1, q5, how='left', left_on='q5', right_on='q5')
b = pd.merge(a, q7, how='left', left_on='q7', right_on='q7')
c = pd.merge(b, q8, how='left', left_on='q8', right_on='Q8')
d = pd.merge(c, q9, how='left', left_on='q9', right_on='Q9')
e = pd.merge(d, q10, how='left', left_on='q10', right_on='q10')
f = pd.merge(e, q11, how='left', left_on='q11', right_on='Q11')
g = pd.merge(f, q14, how='left', left_on='q14', right_on='Q14')
h = pd.merge(g, q16, how='left', left_on='q16', right_on='Q16')
i = pd.merge(h, q17, how='left', left_on='q17', right_on='Q17')
j = pd.merge(i, q18, how='left', left_on='q18', right_on='Q18')
```

```python
#list(j.columns.values)
```

```python
colsToDrop = ['q3','q5','q7','q8','q9','q10','q11','q12','q13','q14','q15','q16','q17','q18','Q3_5','Q8','Q9',
 'Q11','Q14','Q16','Q17','Q18']
j.columns = [c.replace(' ', '_') for c in j.columns]
j.drop(colsToDrop, axis=1)
```

```python
AM = j[j['tim_per']=='1-AM_Peak']
PM = j[j['tim_per']=='3-PM_Peak']
```

```python
AM.groupby('TripPurpose').size()
```

```python
colors = ["#BDD9BF", "#2E4052", "#FFC857", "#9FA2B2", "#412234"]
AM.Destination_Purpose.value_counts().plot(
    kind='pie', 
    colors=colors)
plt.axis('equal')
plt.title('AM Ridership on the 16Y - TRIP PURPOSE')
```

```python
colors = ["#BDD9BF", "#2E4052", "#FFC857", "#9FA2B2", "#412234"]
PM.Destination_Purpose.value_counts().plot(
    kind='pie',
    colors=colors)
plt.axis('equal')
plt.title('PM Ridership on the 16Y - TRIP PURPOSE')
```
